```python
import pandas as pd 
import numpy as np 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
filepath = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv"
df_can = pd.read_csv(filepath)
print('Data read into a pandas dataframe!')
```

```python
df_can.head()
```

```python
df_can.set_index('Country', inplace=True)
```

```python
df_can.head(3)
```

```python
df_can.index.name = None
df_can.head(2)
```

```python
years = list(map(str, range(1980,2014)))
years
```

```python
print('Matplotlib version:', mpl.__version__)
```

```python
print(plt.style.available)
mpl.style.use(['ggplot'])
```

```python
haiti = df_can.loc['Haiti', years]
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')

plt.show()
```

```python
haiti.index.dtype
```

```python
haiti.index = haiti.index.astype(int)
```

```python
haiti.index.dtype
```

```python
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')

y = haiti.values.tolist()
x = haiti.index.to_list()

max_index = np.argmax(y)
max_x = x[max_index]
max_y = y[max_index]

plt.text(max_x, max_y, '2010 Earthquake')

plt.show()
```

```python
df_CI = df_can.loc[['India', 'China'], years]
df_CI
```

```python
df_CI.columns.astype(int)
```

```python
df_CI.plot(kind='line')

plt.title('China & India')
plt.xlabel('Year')
plt.ylabel('Number of immigrants')


plt.show()
```

```python
df_CI = df_CI.transpose()
df_CI.head()
```

```python
df_CI.plot(kind='line')

plt.title('Immigrants from China and India')
plt.xlabel('Year')
plt.ylabel('Number of immigrants')


plt.show()
```

```python
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
```

```python
df_top_5 = df_can.head()
df_top_5
```

```python
df_top_5.loc[:, years]
```

```python
df_top5 = df_top_5.loc[:, years].transpose()
df_top5.head()
```

```python
df_top5.rename({'United Kingdom of Great Britain and Northern Ireland':'UK'}, axis=1, inplace=True)
df_top5.head()
```

```python
df_top5.index.astype(int)
```

```python
df_top5.plot(kind='line', figsize=(14, 8))

plt.title('Top 5 Countries Immigration to Canada')
plt.xlabel('Year')
plt.ylabel('Number of immigrants')

plt.show()
```

```python
df_top5.plot(kind='area', figsize=(14, 8))

plt.title('Top 5 Countries Immigration to Canada')
plt.xlabel('Year')
plt.ylabel('Number of immigrants')

plt.show()
```
