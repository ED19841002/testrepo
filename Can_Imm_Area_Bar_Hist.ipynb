```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
mpl.style.use('ggplot')
```

```python
filepath = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv'
df_can = pd.read_csv(filepath)
print('Data read into a pandas dataframe!')
```

```python
df_can.head()
```

```python
df_can.shape
```

```python
df_can.set_index('Country', inplace=True)
df_can.head()
```

```python
print('data dimensions:', df_can.shape)
```

```python
df_can.index.name = None
```

```python
df_can.head()
```

```python
years = list(map(str, range(1980,2014)))
```

```python
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = df_can.head()
df_top5 = df_top5[years].transpose()
df_top5.head()
```

```python
df_top5.rename({'United Kingdom of Great Britain and Northern Ireland':'UK'}, axis=1, inplace=True)
df_top5.head()
```

```python
df_top5.index.astype(int)
```

```python
#using Scripting layer
df_top5.plot(kind='area', stacked=False, figsize=(20, 10))

plt.title('Immigration Trend of Top 5 Countries')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
df_top5.plot(kind='area',
             alpha=0.25,  # 0 - 1, default value alpha = 0.5
             stacked=False,
             figsize=(20, 10))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
#using artist layer
ax = df_top5.plot(kind='area',
             alpha=0.35,  # 0 - 1, default value alpha = 0.5
             stacked=False,
             figsize=(20, 10))

ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')

plt.show()
```

```python
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)

df_top15 = df_can['Total'].head(15)
df_top15
```

```python
type(df_top15)
```

```python
# generate plot
df_top15.plot(kind='barh', figsize=(12, 12), color='steelblue')
plt.xlabel('Number of Immigrants')
plt.title('Top 15 Conuntries Contributing to the Immigration to Canada between 1980 - 2013')

# annotate value labels to each country
for index, value in enumerate(df_top15):
  label = format(int(value), ',') # format int with commas

# place text at the end of bar (subtracting 47000 from x, and 0.1 from y to make it fit within the bar)
  plt.annotate(label, xy=(value - 47000, index - 0.10), color='white')

plt.show()
```
